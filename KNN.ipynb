```python
data['Loan_Status'] = data['Loan_Status'].apply(lambda x:1 if x=='Y' else 0)
data['Loan_Status'] = data['Loan_Status'].astype('int64')
fig = plt.figure(figsize=(20,20))
i=1
for cols in data[cat_cols]:
    plt.subplot(4,2,i)
    sns.barplot(x=cols, y='Loan_Status',data=data);
    plt.title('Loan Status by {}'.format(cols))
    i=i+1
    
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# target column
tgt_col = ['Loan_Status']

# Categorical cols
category_names = data.nunique()[data.nunique() < 15].keys().tolist()
category_names = [x for x in category_names if x not in tgt_col]

# Numerical cols
num_cols = [i for i in data.columns if i not in category_names + tgt_col]

# Binary cols
bin_cols = data.nunique()[data.nunique()==2].keys().tolist()

# Multi-cols
multi_cols = [i for i in category_names if i not in bin_cols]

# Label Encoding Binary cols
le = LabelEncoder()
for i in bin_cols:
    data[i] = le.fit_transform(data[i])
    
# Duplicating cols for multi-value columns
data = pd.get_dummies(data=data, columns=multi_cols)
```

```python
cont_features = []
for features in data.select_dtypes(include=['int64','float64']):
    if features != 'Loan_Status':
        cont_features.append(features)
```

```python
ld_df = data
# Scaling numerical features
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
ld_df[cont_features] = minmax.fit_transform(ld_df[cont_features].values)
```

```python
cols = [i for i in ld_df.columns if i not in tgt_col]
X = ld_df[cols]
Y = pd.DataFrame(ld_df['Loan_Status'])
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
folds = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
for train_index, test_index in folds.split(X,Y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Setup arrays to store training and test accuracies
neighbors = np.arange(1,9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Fit the model
    knn.fit(X_train, Y_train)
    
    # Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, Y_train)
    
    # Compute accuracy on test set
    test_accuracy[i] = knn.score(X_test, Y_test)
```

```python
plt.title('K-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training Accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
# Fit the model
knn.fit(X_train, Y_train)
# Getting accuracy
knn.score(X_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix
#let us get the predictions using the classifier we had fit above
Y_pred = knn.predict(X_test)
confusion_matrix(Y_test, Y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
Y_pred_proba = knn.predict_proba(X_test)[:,1]
from sklearn.metrics import roc_curve
fpr, tpr, thresholds  = roc_curve(Y_test, Y_pred_proba)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='knn')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('Knn(n_neighbors=5) ROC Curve')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test, Y_pred_proba)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':np.arange(1,50)}
knn=KNeighborsClassifier()
knn_cv = GridSearchCV(knn, param_grid, cv=5)
knn_cv.fit(X,Y)
```
