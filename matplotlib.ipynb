```python
n_groups = 5 
means_a =(30,40,50,60,70)
means_b =(30,90,20,60,80)
means_c =(30,10,20,60,100)
fig, ax= plt.subplots()
index = np.arange (n_groups)
bar_width = 0.5
rects1 = plt.bar(index,means_a,0.05,color = 'red',label = 'a')
rects2 = plt.bar(index + bar_width /4 ,means_b,0.05,color = 'blue',label = 'b')
rects3 = plt.bar(index + bar_width/2,means_c,0.05,color = 'green',label = 'c')

plt.xlabel('class')
plt.ylabel('confidence')
plt.legend() #displaying colors of a,b,c on the plot
plt.tight_layout()
plt.show()
```
